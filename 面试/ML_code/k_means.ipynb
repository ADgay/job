```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def gen_data():
    center = np.array([[1,1], [2,2], [1,2]])
    X1 = np.random.randn(100,2)*0.3+[center[0]]
    X2 = np.random.randn(100,2)*0.5+center[1]
    X3 = np.random.randn(100,2)*0.6+center[2]
    X = np.concatenate([X1,X2,X3], 0)
    print(X.shape)
    return X
```

```python
X = gen_data()
XX = X.copy()
plt.scatter(XX.T[0, 0:100], XX.T[1, 0:100], color = 'blue',alpha=0.5)
plt.scatter(XX.T[0, 100:200], XX.T[1, 100:200], color = 'green',alpha=0.5)
plt.scatter(XX.T[0, 200:], XX.T[1, 200:], color = 'red',alpha=0.5)
```

```python
k = 3
N = X.shape[0]
index = np.zeros(N)*0.0
np.random.shuffle(X)
k_vector = X[:k]
ppp = k_vector.copy()
k_vector_init = np.zeros_like(k_vector)*0.0
print(k_vector_init.shape)
```

```python
count = 0
while(np.mean(np.abs(k_vector-k_vector_init))>0.000001):
    count += 1
    print(count)
    k_vector_init = k_vector.copy()
    for i in range(N):
        index[i] = np.argmin(np.mean(([X[i]]-k_vector_init)**2, -1))
    for j in range(k):
        k_vector[j] = np.mean(X[index==j], 0)
```

```python
plt.scatter(XX.T[0, 0:100], XX.T[1, 0:100], color = 'blue',alpha=0.5)
plt.scatter(XX.T[0, 100:200], XX.T[1, 100:200], color = 'green',alpha=0.5)
plt.scatter(XX.T[0, 200:], XX.T[1, 200:], color = 'red',alpha=0.5)

for i in range(k):
    plt.scatter(k_vector.T[0, i], k_vector.T[1, i], color = 'black',s = 150)
    
for i in range(k):
    plt.scatter(ppp.T[0, i], ppp.T[1, i], color = 'black', marker = '+',s = 150)
```
